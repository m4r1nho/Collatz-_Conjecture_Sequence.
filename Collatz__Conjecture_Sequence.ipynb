# **Collatz _Conjecture_Sequence**

```python
#input the starting number
n = int(input('Type a natural value: '))

# the vector to store the sequence values
vector = []
vector.append(n)

while n>1: # before the loop

  if n % 2 == 0: #even
    n = n/2

  else: #odd
    n = n*3+1

  vector.append(n)
  print(vector)

if n == 1: # first loop
  print(f'Number of interactions : {len(vector)}')
  print(f'The greatest value: {max(vector)}')
```

```python
import matplotlib.pyplot as plt

y = vector # sequence in y
x = range(len(vector))

# create the line plot
plt.plot(x, y)

# add labels and title
plt.xlabel("Value")
plt.ylabel("Iteration")
plt.title("Collatz Conjecture Sequence")

# show the plot
plt.show()
```

```python
def sequence(value, data_history):
    vector = []
    vector.append(value)

    while value > 1:
        if value % 2 == 0:
            value = value / 2
        else:
            value = value * 3 + 1

        vector.append(value)

    data_history.append([vector[0], len(vector), max(vector)])  # Adicionando ao histórico antes de modificar o valor

    return data_history


def test(final_test):
    data_history = []
    for i in range(final_test):
        data_history = sequence(i, data_history)
    return data_history
```

```python
data_history = test(final_test=1000000)
```

```python
import pandas as pd

# names of columns
list_columns = ['Number', 'Interactions', 'Max']

# a pandas dataframe
df = pd.DataFrame(data_history, columns=list_columns)
df
```

```python
# the top 5 values with the highest interactions
df_interactions = df.sort_values(by='Interactions', ascending=False)
print(df_interactions.head(5))
```
